```python
stopwords = ['ì˜/Josa','ê°€/Josa','ì´/Josa','ì€/Josa','ë“¤/Josa','ëŠ”/Josa','ì¢€/Noun','ê±/Noun','ê³¼/Josa','ë„/Josa','ë¥¼/Josa','ìœ¼ë¡œ/Josa','ì—/Josa','ì™€/Josa','í•œ/Noun','í•˜ë‹¤/Verb','ì„/Josa',
             'ì—ì„œ/Josa','ì—ê²Œ/Josa', 'í•˜ê³ /Josa', 'ì´ë‹¤/Verb', './Punctuation']
```

```python
from tqdm.notebook import tqdm

def okt_pos_tokenizer(doc):
    result = []
    
    for sentence in tqdm(doc):
        tokenized_sentence = Okt().pos(sentence, norm=True, stem=True, join=True)
        stopwords_removed_sentence = [word for word in tokenized_sentence if not word in stopwords] # ë¶ˆìš©ì–´ ì œê±°
        result.append(stopwords_removed_sentence)
    return result

df['token'] = okt_pos_tokenizer(df['user_reviews'])
```

```python
# ì‹œë¦¬ì¦ˆë¥¼ ë¦¬ìŠ¤íŠ¸í™”
tmp_token = []

for text in df.token:
    tmp_token += text

tmp_token
```

```python
ko = nltk.Text(tmp_token) 
ko_freq = ko.vocab()
ko_freq
```

```python
threshold = 3
total_cnt = len(ko_freq) # ë‹¨ì–´ì˜ ìˆ˜
rare_cnt = 0 # ë“±ì¥ ë¹ˆë„ìˆ˜ê°€ thresholdë³´ë‹¤ ì‘ì€ ë‹¨ì–´ì˜ ê°œìˆ˜ë¥¼ ì¹´ìš´íŠ¸
total_freq = 0 # ë°ì´í„°ì˜ ì „ì²´ ë‹¨ì–´ ë¹ˆë„ìˆ˜ ì´ í•©
rare_freq = 0 # ë“±ì¥ ë¹ˆë„ìˆ˜ê°€ thresholdë³´ë‹¤ ì‘ì€ ë‹¨ì–´ì˜ ë“±ì¥ ë¹ˆë„ìˆ˜ì˜ ì´ í•©

# ë‹¨ì–´ì™€ ë¹ˆë„ìˆ˜ì˜ ìŒ(pair)ì„ keyì™€ valueë¡œ ë°›ëŠ”ë‹¤.
for key, value in ko_freq.items():
    total_freq = total_freq + value

    # ë‹¨ì–´ì˜ ë“±ì¥ ë¹ˆë„ìˆ˜ê°€ thresholdë³´ë‹¤ ì‘ìœ¼ë©´
    if(value < threshold):
        rare_cnt = rare_cnt + 1
        rare_freq = rare_freq + value

print('ë‹¨ì–´ ì§‘í•©(vocabulary)ì˜ í¬ê¸° :',total_cnt)
print('ë“±ì¥ ë¹ˆë„ê°€ %së²ˆ ì´í•˜ì¸ í¬ê·€ ë‹¨ì–´ì˜ ìˆ˜: %s'%(threshold - 1, rare_cnt))
print("ë‹¨ì–´ ì§‘í•©ì—ì„œ í¬ê·€ ë‹¨ì–´ì˜ ë¹„ìœ¨:", (rare_cnt / total_cnt)*100)
print("ì „ì²´ ë“±ì¥ ë¹ˆë„ì—ì„œ í¬ê·€ ë‹¨ì–´ ë“±ì¥ ë¹ˆë„ ë¹„ìœ¨:", (rare_freq / total_freq)*100)
```

### Vectorizer + MultinomialNB

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
```

```python
df['vect'] = [' '.join(text) for text in df.token]
```

```python
df.head()
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df['vect'], df['label'], test_size=0.2, random_state=13)

# Convert the text data into numerical vectors using TF-IDF
vectorizer = TfidfVectorizer()
X_train_vectorized = vectorizer.fit_transform(X_train)
X_test_vectorized = vectorizer.transform(X_test)

# Train a Naive Bayes classifier
multinomialNB = MultinomialNB()
multinomialNB.fit(X_train_vectorized, y_train)

# Make predictions on the test set
pred = multinomialNB.predict(X_test_vectorized)
pred_proba = multinomialNB.predict_proba(X_test_vectorized)
# Evaluate the model
accuracy = accuracy_score(y_test, pred)
print(f"Accuracy: {accuracy:.2f}")

# Print classification report
print(classification_report(y_test, pred))
```

- ë‹¤ë¥¸ ì ìˆ˜ í™•ì¸í•˜ê¸°
- ë‹¤ì¤‘í´ë˜ìŠ¤ì˜ f1_scoreì˜ í‰ê· 
    - ë¼ë²¨ì˜ ì¤‘ìš”ë„ì— ë”°ë¼ ì–´ë–¤ ì§€í‘œë¥¼ ì„ íƒí• ì§€ê°€ ë‹¬ë¼ì§„ë‹¤. 
    - ëª¨ë“  ë¼ë²¨ì´ ìœ ì‚¬í•œ ì¤‘ìš”ë„ë¥¼ ê°€ì§„ë‹¤ë©´ macro average ê°’ì„ ì°¸ê³ í•˜ê³ 
    - ìƒ˜í”Œì´ ë§ì€ ë¼ë²¨ì— ì¤‘ìš”ë„ë¥¼ ë‘ê³  ì‹¶ë‹¤ë©´ weighted average ê°’ì„ ì°¸ê³ í•˜ë©´ ëœë‹¤
    - ê·¸ë¦¬ê³  ë¼ë²¨ì— ìƒê´€ì—†ì´ ì „ì²´ì ì¸ ì„±ëŠ¥ì„ í‰ê°€í•˜ê³  ì‹¶ì„ ë•ŒëŠ” micro average, ì¦‰ accuracy ë¥¼ í™œìš©í•˜ë©´ ëœë‹¤.

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, roc_auc_score

def get_clf_evalscore(y_test, pred, pred_proba):
    acc = accuracy_score(y_test, pred)
    f1 = f1_score(y_test, pred, average='macro')
    pre = precision_score(y_test, pred, average='macro')
    rec = recall_score(y_test, pred, average='macro')
    # auc = roc_auc_score(y_test, pred_proba[:,1], multi_class='ovr')   #ë‹¤ë¦„. ì£¼ì˜

    return acc, pre, rec, f1

from sklearn.metrics import confusion_matrix

def print_clf_evalmatrix(y_test, pred, pred_proba):
    confusion = confusion_matrix(y_test, pred)
    acc, pre, rec, f1 = get_clf_evalscore(y_test, pred, pred_proba)

    print('==> Confusion Matrix')
    print(confusion)
    print('=====================')
    print('Accuracy: {0:.4f}, Precision: {1:.4f}'.format(acc, pre))
    print('Recall: {0:.4f}, F1: {1:.4f}'.format(rec, f1))  
```

```python
print_clf_evalmatrix(y_test, pred, pred_proba)
```
